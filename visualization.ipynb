```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()

# Load dataset
file_path = 'GOOG.csv
google_stock_data = pd.read_csv(file_path)

# Preprocess
google_stock_data['Date'] = pd.to_datetime(google_stock_data['Date'], format='%d-%m-%Y')
google_stock_data.sort_values('Date', inplace=True)
google_stock_data.set_index('Date', inplace=True)

# Plotting graph
def analyze_feature(data, feature_name):
    # Line plot
    plt.figure(figsize=(14, 7))
    plt.plot(data.index, data[feature_name], label=f'{feature_name} Over Time', color='blue')
    plt.title(f'{feature_name} Over Time')
    plt.xlabel('Year')
    plt.ylabel(feature_name)
    plt.legend()
    plt.grid(True)
    plt.show()

    # Boxplot
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data[feature_name], color='purple')
    plt.title(f'Boxplot of {feature_name}')
    plt.xlabel(feature_name)
    plt.show()

    # Distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(data[feature_name], kde=True, bins=30, color='green')
    plt.title(f'Distribution of {feature_name}')
    plt.xlabel(feature_name)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

features = ['Low', 'Open', 'Volume', 'High', 'Close', 'Adjusted Close']

for feature in features:
    analyze_feature(google_stock_data, feature)

# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = google_stock_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
